# Exercício 1

Uma Series (array unidimensional) da biblioteca Pandas pode ser criada utilizando-se o construtor:

* pd.Series( data, index, dtype, copy )

Então, podemos criar objetos do tipo Series a partir de array, dicionários, escalares e iteráveis. Com base nessas informações responda as questões a seguir.



a) A partir da lista abaixo crie um objeto do tipo Series.

```
lista = ['a', 'b', 'c', 'd']
```

Resultado esperado:

```
0    a
1    b
2    c
3    d
dtype: object
```

b) A partir da lista anterior obtenha:

```
10    a
11    b
12    c
13    d
dtype: object
```

Dica: mude o argumento index.


c) Obtenha a Series abaixo:

```
0    10
1    10
2    10
3    10
dtype: int64
```

d) A partir do array,

```
import numpy as np
array = np.arange(10,20,2)
```

Obtenha:

```
1    10.0
2    12.0
3    14.0
4    16.0
5    18.0
dtype: float64
```

````python
# a)
lista = ['a', 'b', 'c', 'd']
a = pd.Series(lista)
print(a)
print()

# b)
b = pd.Series(data=lista, index=(range(10,14)))
print(b)
print()

# c)
c = pd.Series(data=np.full(fill_value=10, shape=4))
print(c)
print()

# d)
d = pd.Series(data=np.arange(start=10, stop=20, step=2), index=(range(1,6)), dtype=float)
print(d)
print()
````
